##  <b id="b1"> Research Question</b>
Whether the Taylor rule is applicable in the low interest era? Does it fit well with the recent Japanese case or it requires some modification in the form of the augmented Taylor rule? 

## <b id="b2"> Motivation</b>

However, in recent times, many countries like Japan have abandoned traditional monetary policy based on interest rates in favour of unconventional monetary policy. In this assignment we try to investigate whether the Taylor rule is still applicable for the Japanese scenario. With some other variables added to the Taylor rule we try to emperically examine the augmented Taylor rule as well. ## <b id="b3"> Methodology</b>
We first collect data of the Interest rate (money market rate), exchange rate, consumer price rate, share price and output gap during the period 2006-16. The price gap/inflation gap is calculated by substracting the consumer price rate with the target level of inflation. As [Bask and Selander (2009)](https://www.researchgate.net/publication/285589504_Robust_Taylor_Rules_under_Heterogeneity_in_Currency_Trade) suggest, we include exchange rates in the Taylor rule due to the heterogeneity in currency trade in the economy. We also think Stock prices are an important factor because [Rigbon and Sack (2003)](http://web.mit.edu/rigobon/www/Robertos_Web_Page/mac_-_rmp_files/rmp.pdf) and [Fuhrer and Tootell (2008)](https://papers.ssrn.com/sol3/papers.cfm?abstract_id=887922) found that equity prices impact policy interest rates. We then run two regressions to see the impact of various factors on the interest rates by estimating the Taylor theorem and augmented Taylor theorem by both the OLS estimation and robust estimation methods.

# <b id="b4"> Introduction</b>

### Monetary Policy Rule
The Taylor rule states that central banks should keep in mind the inflation rate and output when setting the monetary policy. Many researchers concluded that monetary policy is conducted in line with the Taylor rule. One of the main advantages of setting the monetary policy according to the Taylor rule is the stability of the economy in combination with transparency and credibility. The policy rule has been examined for many countries in Europe. However, economic literature regarding the low or zero interest rate policy in Japan has been scarce.

### Japanese Experience since the '80s
During the 1980s the economic growth in Japan was relatively high compared to other countries. Moreover, stock and land prices were rapidly increasing, and the Japanese Yen appreciated immensely. In most cases, this would lead to a decrease in export, however, the economy including exports were not affected. However, in the late 1980s the economic bubble bursted which led to a severe recession in 1990s and early 2000. <br/><br/>
In 1999 The Bank of Japan started a new monetary policy to tackle the low or negative economic growth, namely, zero interest policy. However, the impact of the zero interest rate policy did not lead to the desired outcome, therefore, The Bank of Japan implemented the monetary easing policy in 2001. In 2006 the quantitative easing policy ended due to signs that deflation was ending and that Japan was coming out of the recession.  <br/><br/>
After the Subprime shock and the fall of Lehman bank, Japan experienced a huge inflow of money into the financial markets. This led to an appreciation of the yen, which damaged the Japanese economy. In 2010 The Bank of Japan again started a program of monetary easing to tackle low economic growth and deflation. At the end of 2010 The Bank of Japan implied the zero interest rate policy, which is still on till now. In 2013, a complete different policy was implemented to tackle deflation and low GDP growth, namely, Abenonomics. This program did not only focus on fiscal and monetary policy, but also on structural reforms. For instance, changes in tax systems, enhancing foundation for innovation and development. <br/><br/>
[Olson and Wohar (2015)](https://pdfs.semanticscholar.org/e77c/8043ea45bcbd1c760a844f3f2da3cb58923d.pdf) proofed that the implementation of a simple Taylor rule could have enhanced the monetary performance in Germany. [Hove (2012)](https://www.djoef-forlag.dk/sites/nt/files/2012/2012_2.pdf) explained that the Taylor rule is different from time to time. [Ruhal (2015)](http://link.springer.com/article/10.1007/s00181-014-0820-z) showed that the ECB policy is less stable than the German Bundesbank, which violated the Taylor rule. There have been many papers in the economic literature regarding this topic, however, the conclusions are different. We take into account the Japanese economic situation and estimate an augmented Taylor rule and investigate if the Japanese monetary policy is in line with the Taylor rule. 

## <b id="b5"> Assumptions</b>
* We assume that the price level is a non-monetary phenomenon i.e. inflation in an economy is caused due to shocks that are exogenous to central bank. The central bank can control inflation but they do not create it. 
If price level is assumed to be a monetary phenomenon, it would mean that the central bank can control inflation. This implies that inflation is caused by a single factor i.e. excess money creation by the central bank. 
In nonmonetary models of inflation there is no single explanation for inflation. Instead a variety of different factors cause inflation.
* The output gap is assumed to be measured by output relative to an exponential trend.
* We have assumed p* (target inflation) as 2%, which was set by the Bank of Japan in 2014.


## <b id="b6"> Empirical Analysis</b>
We emperically analyze the Taylor rule and augmented Taylor rule in this section.<br/>

The traditional Taylor rule can be given by:<br/> $$i_t = \alpha + \beta \pi_t + \gamma y_t + \epsilon_t$$

Where,<br/>
$i_t$ denotes target interest rate or the overnight money market rate.  
$\pi_t$ = $p-p*$, p is inflation or the consumer price rate. p* is the target inflation rate, which was set by the BOJ (Bank Of Japan).<br/>
$y_t$ is the output gap. 



Roskelley(2016) in his paper argues that central bank should not blindly follow a simple rule. If necessary variables are not included along with the Taylor rule in setting monetary policy then it may lead to large deviation from estimated policy rule. Therefore we also study the augmented Taylor rule. Along with price and output gap we have added two more variables in our model i.e. stock price (NIKKIE 225) and effective exchange rate. 

The augmented Taylor can be given by:
$$i_t = \alpha + \beta \pi_t + \gamma y_t + \varphi X_t + \epsilon_t$$

Where,<br/>
$X_t$ denotes a vector of other variables i.e. stock price and effective exchange rate.

When we extend this model to incorporate the price of the Nikkie 225 index and the exchange rate of the Yen, the adjusted R-squared jumps to 57.6%. The output gap, exchange rate and price of the index are statistically significant at the 99% confiednce interval, while the price gap in this model becomes significant at the 95% confidence interval.

Least square estimation models are highly sensitive to outliers or extreme observations. It is not a problem if the outlier is simply an extreme observation drawn from the tail of a normal distribution, however, if the outlier is from non-normal measurement error or some other violation of standard OLS, it compromises the validity of the regression results if a non-robust technique is employed. In our data the sample size is small, along with some outliers that do not follow the pattern of the other observations. Robust estimation (unlike maximum likelihood estimation) is not sensitive to the observation that do not follow the pattern of the other observation. Therefore we decided to use robust estimation along with the usual OLS.

```python
res=sm.RLM(y,X).fit()
print(res.summary())
```

```python
res1=sm.RLM(y,X1).fit()
print(res1.summary())
```
